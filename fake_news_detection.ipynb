```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
#80% training data and 20% testing data
from sklearn.feature_extraction.text import TfidfVectorizer
#term freq(tf)-inverse document freq(idf)
#tf is used for no of times the term appears in a document
#idf for metric calculate log and division
from sklearn.svm import LinearSVC
#if we use randome forest classifier then it give a bad result
```

```python
#load the data
data=pd.read_csv("fake_or_real_news.csv")
```

```python
data
```

```python
#if fake then then return 1 else otherwise 0 that is store in the fake column
data['fake']=data['label'].apply(lambda x:0 if x=="REAL" else 1)
```

```python
#after creating new cloumn based on the label then we are print
data
```

```python
#delete the label column from the table
data=data.drop('label',axis=1)
```

```python
data
#text and fake columns as take features for good accuracy
```

```python
#text is feature and fake is target variable I am consider
X=data['text']
y=data['fake']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
#20% for evaltion and 80% for training data so test_size 20
```

```python
len(X)
```

```python
len(X_train) #(len(X)*80/100)
```

```python
len(X_test) #(len(X)*20/100)
```

```python
#if we use linear SVC then it must do Vectorization
#X is sentences we want numbers so use vectorize
vectorize = TfidfVectorizer(stop_words='english', max_df=0.7)
X_train_vectorized = vectorize.fit_transform(X_train)
X_test_vectorized = vectorize.transform(X_test)
```

```python
clf=LinearSVC()
clf.fit(X_train_vectorized,y_train) #y_train is numbered so we can not do vectorizer
```

```python
#accuracy by score
clf.score(X_test_vectorized,y_test)  #0.9486 near to 95% accuracy on testing side
```
